## 1. noise recover

If use percentage exp(-1/20): 95% original picture plus 5% noise each time. Training process assume markov property.i.e. The distance between step i and step i-1 image is independent of previous image. Each time noise are generated randomly. We first try step of 50.

```python
main_2(contact = True, col=False, overlap=True,m=8,stride = 4,epoches=3,sub=True,step=50)
```

When reach step of 15, the minimization of loss can't well work.

```python
main_2(contact = True, col=False, overlap=True,m=8,stride = 4,epoches=3,sub=True,step=15)
```

```python
main_2(contact = True, col=False, overlap=True,m=8,stride = 4,epoches=5,sub=True,step=12)
```

If we try to increase step distance, the result seems to be better, which contradict to our assumption.

```python
main_2(contact = True, col=False, overlap=True,m=8,stride = 4,epoches=3,sub=True,step=5)
```

```python
main_2(contact = True, col=False, overlap=True,m=8,stride = 4,epoches=3,sub=True,step=12)
```

Might exist a critical point C(function of step length) where noise count certain precentage, the model can't work well (after taht point, loss increase geometrically).
